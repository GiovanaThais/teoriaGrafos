```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (9, 5)
plt.rcParams['figure.facecolor'] = '#00000000'
```

### Preparação de dados : envolve limpeza dos dados, transformação de formatos, tratamento de missing values

```python
#Carregando Dataframe escolhido no Kaggle
df=pd.read_csv('agricultural_raw_material.csv')

#Checando os tipos de variáveis das colunas
df.dtypes
```

```python
df.info
#Checando os valores NULL
df.isnull().sum()
```

```python
#Tratamento dos dados com replacing
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)

#Remover as tabelas com valor NaN
df = df.dropna()

#Checar a remoção
df.isnull().sum()

#Formatando tipos de dados para float
lst = ["Coarse wool Price", "Coarse wool price % Change", "Copra Price", "Copra price % Change", "Cotton price % Change","Fine wool Price", "Fine wool price % Change", "Hard log price % Change", "Hard sawnwood price % Change", "Hide price % change", "Plywood price % Change", "Rubber price % Change", "Softlog price % Change", "Soft sawnwood price % Change", "Wood pulp price % Change"]
df[lst] = df[lst].astype("float")
df.dtypes
```

```python
df.head()
```

```python
#Formatando a data e transformando em index
df.Month = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
df = df.set_index('Month')
```

```python
df.head()
```

```python
#preço das matérias sem o a variação percentual
materia_prima = df[['Coarse wool Price',  'Copra Price','Cotton Price', 'Fine wool Price',  'Hard log Price', 'Hard sawnwood Price',
 'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']]

materia_prima.head()
```

```python
preco_mudanca=df[['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change',
            'Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change',
            'Soft sawnwood price % Change','Wood pulp price % Change']]

preco_mudanca.head()
```

```python
print(f'Amplitude dos preços: \n{materia_prima.max() - materia_prima.min()}')
print(f'\nAmplitude da variação dos preços: \n{preco_mudanca.max() - preco_mudanca.min()}')
```

```python
#Gráfico que denota variação do preço de produtos semelhantes ao longo dos anos
plt.figure(figsize=(10, 10))
plt.plot(df[['Coarse wool Price',  'Fine wool Price', 'Rubber Price', 'Hide Price','Hard sawnwood Price']])
plt.title("Comparação preço da matéria prima")
plt.xlabel('Anos')
plt.ylabel('Preços')
plt.legend(['Preço da Lã grossa','Preço da Lã Fina', 'Preço da Borracha','Preço das Peles','Preço da Madeira serrada dura'],loc='upper center',bbox_to_anchor=(0.3,1))
```

```python
#Agrupando os dados pelo ano e calculando média'
media_ano = materia_prima.groupby('Month')[['Coarse wool Price',  'Fine wool Price', 'Rubber Price', 'Hide Price','Hard sawnwood Price']].sum().mean()

# A variável 'media_por_ano' agora contém a média dos preços por ano
print(media_ano)
```

```python
moda_ano = materia_prima[['Coarse wool Price',  'Fine wool Price', 'Rubber Price', 'Hide Price','Hard sawnwood Price']].mode()

# Exibindo a Moda
print(moda_ano)
```

```python
mediana_ano = materia_prima.median()

# Exibindo a Mediana
print(mediana_ano)
```

```python
1#Calcular a variância
variancia_ano = preco_mudanca.var()

variancia_ano
```

```python
# Calcular o desvio padrão
desvio_ano = materia_prima.std()

desvio_ano
```

```python
#Interquartil
quartil = materia_prima['Coarse wool Price'].quantile([0.25, 0.75])
iqr = quartil[0.75] - quartil[0.25]
print(iqr)
```

### Visualização dos resultados:gráficos de barras, gráficos de dispersão, histogramas.

```python
#Histograma
materia_prima.hist(figsize=(11,14))
```

```python
preco_mudanca[['Fine wool price % Change', 'Rubber price % Change','Hard sawnwood price % Change']].plot.density()
plt.title('Gráfico de densidade')
plt.legend(['Alteração preço Lã Fina','Alteração preço Borracha','Alteração preço Madeira serrada dura'],loc='upper center',bbox_to_anchor=(0.75,1.02))
plt.show()
```

```python
#Gráfico de dispersão que denota  variação de preços de determinados produtos ao longo dos anos
lowlist=['Coarse wool Price',  'Fine wool Price', 'Rubber Price', 'Hide Price','Hard sawnwood Price']
plt.figure(figsize=(12,12))
plt.ylabel('Prices')
plt.xlabel('Years')
for i in range(len(lowlist)):
    sns.scatterplot(y=df[lowlist[i]],x=df.index);
    plt.legend(['Preço da Lã grossa','Preço da Lã Fina', 'Preço da Borracha','Preço das Peles','Preço da Madeira serrada dura'],loc='upper center',bbox_to_anchor=(0.4,1))
```

```python
# variação normal do preço de cada matéria-prima
changelist= ['Copra price % Change','Softlog price % Change','Rubber price % Change',
               'Cotton price % Change','Coarse wool price % Change','Fine wool price % Change',
               'Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change',
               'Soft sawnwood price % Change','Wood pulp price % Change']
for i in range(len(changelist)):
    plt.figure(figsize=(8,4))
    df[changelist[i]].hist(figsize=(6, 3), linewidth=1)
    plt.xlabel('% Change')
    plt.ylabel('count')
    plt.legend(changelist[i:],loc='upper center')
```
